```python
from melo.api import TTS
# Speed is adjustable

# CPU is sufficient for real-time inference.
# You can set it manually to 'cpu' or 'cuda' or 'cuda:0' or 'mps'
device = 'auto' # Will automatically use GPU if available
model = TTS(language='EN', device=device)
speaker_ids = model.hps.data.spk2id
```

```python
speaker_ids
```

```python
def text2speech(text,speed,speaker_id,output_path="./output.wav"):
    model.tts_to_file(text, speaker_ids[speaker_id], output_path, speed=speed)
    return output_path
```

```python
from playsound import playsound

def play_audio(wav_path):
    try:
        playsound(wav_path)
    except playsound.PlaysoundException as e:
        print(f"Error playing audio: {e}")
        print("Possible solutions:")
        print("1. Ensure the audio file exists and is a valid format.")
        print("2. Check audio device configuration and drivers.")
        print("3. Restart your runtime environment.")
```

```python
import simpleaudio as sa
import os

file_path = "/content/gdrive/MyDrive/MeloTTS/MeloTTS/output.wav"


# Function to play audio, handling potential errors
def play_audio(file_path):
    try:
        wave_obj = sa.WaveObject.from_wave_file(file_path)
        play_obj = wave_obj.play()
        play_obj.wait_done()
    except sa.SimpleaudioError as e:
        print(f"Error playing audio: {e}")
        print("Possible solutions:")
        print("1. Ensure you have PulseAudio installed: Run '!apt-get update && apt-get install -y pulseaudio'")
        print("2. Check audio device configuration and drivers.")
        print("3. Restart your runtime environment.")
```
